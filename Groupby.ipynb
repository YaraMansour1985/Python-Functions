```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/AXSOSACADEMY/01-Fundamentals/Week02/Data/insurance_mod.csv')

df
Age_Columns=df['age']
Age_Columns.mean()
```

```python
Charges_columns=df['charges']
Charges_columns.mean()
```

```python
charges_by_sex=df.groupby('sex')['charges'].mean()
charges_by_sex
```

```python
charges_by_smoker=df.groupby('smoker')['charges'].mean()
charges_by_smoker
```

```python
charges_by_smoker_sex=df.groupby(['smoker','sex'])['charges'].mean()
charges_by_smoker_sex
```
